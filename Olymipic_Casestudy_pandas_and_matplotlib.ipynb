**CASE STUDY ON OLYMPICS OVER THE PAST FEW YEARS..**

```python
import pandas as pd
df = pd.read_excel('Olympic Athletes.xlsx')
print(df)
```

**1) How many Medals did each player won?**

```python
x = df.groupby('Athlete')
print(x['Total Medals'].sum())
```

**2) All the players belonging to Australia**

```python
x = df.loc[(df['Country'] == 'Australia')]
print(x[['Athlete','Age','Sport','Year','Total Medals']])
```

 **3) All the players whose age is 17 and participated in swimming in the year 2008**

```python
x = df.loc[(df['Year'] == 2008) & (df['Sport'] == 'Swimming') & (df['Age'] == 17.0)]
print(x[['Athlete','Country','Total Medals']])
```

**4)Data Log of Usain Bolt**

```python
x = df.loc[(df['Athlete'] == 'Usain Bolt')]
print(x[['Athlete','Year','Sport','Gold Medals','Silver Medals','Bronze Medals','Total Medals']])
```

**5)Number of athletes from each country in the year 2000**

```python
x = df[df['Year'] == 2000].groupby('Country')['Athlete'].count()
print(x)
```

**6) Dominating Country in each sport**

```python
x = df.groupby(['Country','Sport']).sum()
dominating_country = x.groupby(level='Sport')['Total Medals'].idxmax()

for sport, country in dominating_country.items():
 print(f"The Country that dominated {sport} is {country[0]}.")
```

**7) Top 5 countries holding the most medals**

```python
x = df.groupby('Country').sum().sort_values('Total Medals',ascending = False)
print(x[['Total Medals']].head())
```

**8) Country which won the most medals in each year.**

```python
x = df.groupby(['Country','Year']).sum()
Highest_year = x.groupby(level='Year')['Total Medals'].idxmax()

for Year, Country in Highest_year.items():
 print(f"The Country that won most medals in {Year} is {Country[0]}.")
```

**9) Athlete who won Least Medals in each sport**

```python
x = df.groupby(['Athlete','Sport']).sum()
least_scored = x.groupby(level = 'Sport')['Total Medals'].idxmin()
for Sport,Athlete in least_scored.items():
  print(f'Least Medals Won in {Sport} is {Athlete[0]}')
```

 **10) Data of Athletes belonging to Great Britan and name starts with D**

```python
x = df.loc[(df['Country'] == 'Great Britain') & (df['Athlete'].str.startswith('D'))]
print(x[['Athlete','Country','Age','Sport','Total Medals']])
```

**The Age Distribution of the Athletes**

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'], bins=20, color='red', edgecolor='black')
plt.title('Distribution of Athletes Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

**Medals won by Top 5 Countries**

```python
import matplotlib.pyplot as plt
top_countries = df.groupby('Country')[['Gold Medals', 'Silver Medals', 'Bronze Medals']].sum().sum(axis=1).nlargest(5)
top_countries_data = df[df['Country'].isin(top_countries.index)]

plt.figure(figsize=(12, 8))
top_countries_data.groupby('Country')[['Gold Medals', 'Silver Medals', 'Bronze Medals']].sum().plot(kind='bar', stacked=True)
plt.title('Comparison of Gold, Silver, and Bronze Medals by Top 5 Countries')
plt.xlabel('Country')
plt.ylabel('Total Medals')
plt.xticks(rotation=45)
plt.legend(title='Medal Type')
plt.show()
```

**Total Medals Won by the top 30 countries**

```python
import matplotlib.pyplot as plt
top_countries = df.groupby('Country')['Total Medals'].sum().nlargest(30)
top_countries.plot(kind='pie', figsize=(10, 6))
plt.title('Total Medals Won by Top 30 Countries')
plt.xlabel('Country')
plt.ylabel('Total Medals')
plt.show()
```

**Sports played in each year**

```python
import matplotlib.pyplot as plt

df['Year'] = pd.to_datetime(df['Year'], format='%Y')

df['Year'] = df['Year'].dt.year

sports_count_per_year = df.groupby('Year')['Sport'].nunique()
#plt.figure(figsize=(15, 9).)
plt.figure(figsize=(12, 6))
sports_count_per_year.plot(kind='bar', color='k')
plt.title('Number of Unique Sports Played in Each Year')
plt.xlabel('Year')
plt.ylabel('Count of Sports')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

**Athletes age distribution**

```python
import matplotlib.pyplot as plt
indices = range(len(df))
plt.figure(figsize=(20, 6))
plt.bar(indices, df['Age'], color='skyblue')
plt.title('Distribution of Athletes\' Ages')
plt.xlabel('Index')
plt.ylabel('Age')
plt.grid(axis='y')  # Grid only on the y-axis
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
age_min = 20
age_max = 40
filtered_df = df[(df['Age'] >= age_min) & (df['Age'] <= age_max)]
ages = filtered_df['Age']
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.title(f'Age Distribution of Athletes (Age {age_min}-{age_max})')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

**Medals of selected countries**

```python
import matplotlib.pyplot as plt
specific_countries = ['United States', 'Russia', 'China', 'United Kingdom']
specific_countries_df = df[df['Country'].isin(specific_countries)]
plt.figure(figsize=(10, 6))
specific_countries_df.pivot_table(index='Year', columns='Country', values='Total Medals', aggfunc='sum').plot(marker='o')
plt.title('Total Medals Over Time for Specific Countries')
plt.xlabel('Year')
plt.ylabel('Total Medals')
plt.grid(True)
plt.show()
```

**Top 10 Athletes with their number of medals**

```python
import matplotlib.pyplot as plt
athlete_medals = df.groupby('Athlete')['Total Medals'].sum().nlargest(10)
colors = ['salmon', 'skyblue', 'lightgreen', 'gold', 'orange', 'pink', 'cyan', 'purple', 'brown', 'lime']
plt.figure(figsize=(10, 6))
athlete_medals.plot(kind='barh', color=colors)
plt.title('Top 10 Athletes with the Most Medals')
plt.xlabel('Total Medals')
plt.ylabel('Athlete')
plt.gca().invert_yaxis()
plt.show()
```

**For a specific year medals distribution in different sports**

```python
import matplotlib.pyplot as plt
year = 2008
sports_year = df[df['Year'] == year]['Sport'].unique()
total_medals_by_sport_year = df[df['Year'] == year].groupby('Sport')['Total Medals'].sum()
plt.figure(figsize=(10, 8))
plt.pie(total_medals_by_sport_year, labels=sports_year, autopct='%1.1f%%', startangle=140)
plt.title(f'Distribution of Medals Across Different Sports in {year}')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
top_countries = df.groupby('Country')['Total Medals'].sum().nlargest(10)
max_medals_country=top_countries.idxmax()
labels = top_countries.index.tolist()
sizes = top_countries.values.tolist()
explode = [0.3 if Country == max_medals_country else 0 for Country in labels]
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=140,shadow='True')
plt.title("Country with the  most medals of the top 10")
plt.axis('equal')
plt.show()
```
